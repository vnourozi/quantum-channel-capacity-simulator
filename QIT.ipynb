```python
#!/usr/bin/env python3
"""
Enhanced Smooth Curve Simulation with Combined Global and Local Optimization
for Quantum Channel Capacity

This script computes and plots a smooth curve of the maximum (optimized over input state parameters)
one-shot coherent information as a function of a channel parameter.
For each channel parameter value, a two-step optimization is performed:
1. A global search using differential_evolution over the Bloch sphere angles (θ and φ).
2. A local refinement using Nelder–Mead starting from the global result.
Spline interpolation is then used to plot a smooth curve.
This approach is intended to capture even small positive optimum values,
consistent with theoretical predictions (e.g., for the amplitude damping channel at low γ).

Author: Vahid Nourozi

References:
  - J. ur Rehman et al., Phys. Rev. A 105, 032616 (2022).
  - Kianvash et al., "Bounding the quantum capacity with flagged extensions", Quantum (2022).
  - I. Devetak, IEEE Trans. Inf. Theory 51, 44 (2005).
  - M. Wilde, "Quantum Information Theory", Cambridge Univ. Press (2017).
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import differential_evolution, minimize
from scipy.interpolate import make_interp_spline
from numpy import linalg as LA

# ----------------------------
# Utility Functions
# ----------------------------

def von_Neumann_entropy(rho, base=2, tol=1e-12):
    """Compute the von Neumann entropy of a density matrix rho using eigvalsh for stability."""
    eigvals = LA.eigvalsh(rho)
    eigvals = np.maximum(eigvals, tol)
    return -np.sum(eigvals * np.log(eigvals) / np.log(base))

def apply_channel(rho, kraus_ops):
    """Apply a quantum channel to state rho using the given Kraus operators."""
    d_out = kraus_ops[0].shape[0]
    out = np.zeros((d_out, d_out), dtype=complex)
    for K in kraus_ops:
        out += K @ rho @ K.conj().T
    return out

def apply_complementary_channel(rho, kraus_ops):
    """
    Compute the complementary channel output.
    The (i,j) element is Tr(K_j rho K_i†).
    """
    d_env = len(kraus_ops)
    rho_env = np.zeros((d_env, d_env), dtype=complex)
    for i in range(d_env):
        for j in range(d_env):
            rho_env[i, j] = np.trace(kraus_ops[j] @ rho @ kraus_ops[i].conj().T)
    return rho_env

def coherent_information(rho, kraus_ops):
    """
    Compute the one-shot coherent information:
      I_c = S(N(ρ)) - S(N^c(ρ))
    """
    rho_out = apply_channel(rho, kraus_ops)
    rho_env = apply_complementary_channel(rho, kraus_ops)
    return von_Neumann_entropy(rho_out) - von_Neumann_entropy(rho_env)

def pure_state_density(theta, phi):
    """
    Construct a pure qubit state:
      |ψ⟩ = cos(θ/2)|0⟩ + exp(iφ) sin(θ/2)|1⟩
    and return its density matrix.
    """
    psi = np.array([np.cos(theta/2), np.exp(1j * phi)*np.sin(theta/2)])
    return np.outer(psi, np.conj(psi))

# ----------------------------
# Channel Definitions
# ----------------------------

def amplitude_damping_kraus(gamma):
    """
    Return Kraus operators for the amplitude damping channel with damping parameter γ.
    K0 = |0⟩⟨0| + sqrt(1-γ)|1⟩⟨1|, K1 = sqrt(γ)|0⟩⟨1|
    """
    K0 = np.array([[1, 0],
                   [0, np.sqrt(1 - gamma)]])
    K1 = np.array([[0, np.sqrt(gamma)],
                   [0, 0]])
    return [K0, K1]

def depolarizing_kraus(p):
    """
    Return Kraus operators for the qubit depolarizing channel:
      D_p(ρ) = (1-p)ρ + (p/3)*(XρX + YρY + ZρZ)
    """
    I = np.eye(2)
    X = np.array([[0, 1], [1, 0]])
    Y = np.array([[0, -1j], [1j, 0]])
    Z = np.array([[1, 0], [0, -1]])
    K0 = np.sqrt(1-p) * I
    K1 = np.sqrt(p/3) * X
    K2 = np.sqrt(p/3) * Y
    K3 = np.sqrt(p/3) * Z
    return [K0, K1, K2, K3]

def flagged_depolarizing_kraus(p):
    """
    Return flagged Kraus operators for a depolarizing channel.
    The channel maps a qubit (C^2) to a flagged state in C^2 ⊗ C^2 (4-dimensional)
    by attaching a flag vector.
    
    L0 = sqrt(1-p) * kron(I, flag0)
    L1 = sqrt(p/3) * kron(X, flag1), etc.
    """
    I = np.eye(2)
    X = np.array([[0, 1], [1, 0]])
    Y = np.array([[0, -1j], [1j, 0]])
    Z = np.array([[1, 0], [0, -1]])
    flag0 = np.array([[1], [0]])
    flag1 = np.array([[0], [1]])
    L0 = np.kron(I, flag0)
    L1 = np.kron(X, flag1)
    L2 = np.kron(Y, flag1)
    L3 = np.kron(Z, flag1)
    L0 = np.sqrt(1-p) * L0
    L1 = np.sqrt(p/3) * L1
    L2 = np.sqrt(p/3) * L2
    L3 = np.sqrt(p/3) * L3
    return [L0, L1, L2, L3]

# ----------------------------
# Combined Global + Local Optimization
# ----------------------------

def optimize_coherent_info_refined(channel_func, param):
    """
    Optimize over input state parameters [θ, φ] to find the maximum coherent information
    for a given channel parameter using differential evolution followed by Nelder-Mead refinement.
    Returns the maximum I_c and the optimal [θ, φ].
    """
    kraus_ops = channel_func(param)
    
    # Objective function: negative coherent information
    def neg_coherent_info(x):
        theta, phi = x
        rho = pure_state_density(theta, phi)
        return -coherent_information(rho, kraus_ops)
    
    bounds = [(0, np.pi), (0, 2*np.pi)]
    # Global search with differential evolution:
    result_de = differential_evolution(neg_coherent_info, bounds, strategy='best1bin', tol=1e-6, polish=True)
    # Use the DE result as the initial guess for local refinement:
    result_local = minimize(neg_coherent_info, result_de.x, method='Nelder-Mead', options={'xtol':1e-8, 'ftol':1e-8})
    # Return refined optimum:
    opt_val = -result_local.fun
    opt_params = result_local.x
    return opt_val, opt_params

def optimal_coherent_info_vs_param(channel_func, param_range):
    """
    For a range of channel parameter values, use the refined optimizer to find the maximum coherent information.
    Returns arrays of channel parameters and corresponding optimal I_c and optimal input parameters.
    """
    max_Ic_vals = []
    opt_params_list = []
    for param in param_range:
        Ic_opt, x_opt = optimize_coherent_info_refined(channel_func, param)
        max_Ic_vals.append(Ic_opt)
        opt_params_list.append(x_opt)
        print(f"Param = {param:.4f}, Max I_c = {Ic_opt:.4f} bits, Optimal (θ, φ) = ({x_opt[0]:.4f}, {x_opt[1]:.4f})")
    return param_range, np.array(max_Ic_vals), np.array(opt_params_list)

def plot_smooth_curve(x, y, xlabel="Channel Parameter", ylabel="Max Coherent Information (bits)",
                      title="Optimal Coherent Information vs Channel Parameter", label=""):
    x_smooth = np.linspace(x.min(), x.max(), 1000)
    spline = make_interp_spline(x, y, k=3)
    y_smooth = spline(x_smooth)
    
    plt.figure(figsize=(8, 5))
    plt.plot(x_smooth, y_smooth, color="blue", linewidth=2, label=label)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.title(title, fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(True)
    plt.show()

def simulate_optimal_curve(channel_func, param_min, param_max, num_points, channel_label, param_name):
    param_range = np.linspace(param_min, param_max, num_points)
    x, y, opt_params = optimal_coherent_info_vs_param(channel_func, param_range)
    plot_smooth_curve(x, y, xlabel=param_name, ylabel="Max Coherent Information (bits)",
                      title=f"Optimal Coherent Information vs {param_name} ({channel_label})", label=channel_label)
    idx_opt = np.argmax(y)
    print(f"\nOver the range of {param_name}, the maximum coherent information for {channel_label} is {y[idx_opt]:.4f} bits at {param_name} = {x[idx_opt]:.4f}")
    return x, y, opt_params

# ----------------------------
# Main Simulation Routine
# ----------------------------
def main():
    # For amplitude damping channel, sweep γ from 0.001 to 0.02 (very low noise)
    print("Simulating optimal coherent information vs. γ for Amplitude Damping Channel")
    simulate_optimal_curve(amplitude_damping_kraus, param_min=0.001, param_max=0.6, num_points=10000,
                             channel_label="Amplitude Damping", param_name=r"$\gamma$")
    
    # For depolarizing channel, sweep p from 0.001 to 0.02.
    print("\nSimulating optimal coherent information vs. p for Depolarizing Channel")
    simulate_optimal_curve(depolarizing_kraus, param_min=0.001, param_max=0.6, num_points=10000,
                             channel_label="Depolarizing", param_name="p")
    
    # For flagged depolarizing channel, sweep p from 0.001 to 0.02.
    print("\nSimulating optimal coherent information vs. p for Flagged Depolarizing Channel")
    simulate_optimal_curve(flagged_depolarizing_kraus, param_min=0.001, param_max=0.6, num_points=10000,
                             channel_label="Flagged Depolarizing", param_name="p")

if __name__ == "__main__":
    main()
```
